**Нормализация признаков с помощью линейного преобразования:**



```python
data_normal = (data1 - data1.min(axis=0)) / (data1.max(axis=0) - data1.min(axis=0))
data_normal.describe()
```

**Разбиение данных на обучающую и тестовую выборки**

```python

X = data_normal.drop('stroke', axis = 1)
y = data_normal['stroke']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=1)
```

Создадим функцию для удобного подбора параметров

```python

def cv_params(model, param_grid):
    scoring = 'neg_mean_absolute_error'
    opt_params = GridSearchCV(
      estimator = model, 
      param_grid = param_grid,
      scoring = scoring, 
      cv = 5,
      n_jobs = -1 ) 
    opt_params.fit(X_train,y_train)
    params = opt_params.best_params_
    best_score = opt_params.best_score_

    print(f'Best score:{round(-best_score,2)}')
    print(f'Best parameters:{params}\n')

    return params
  
```

Параметры для KNN




```python


knn_param_grid = {
    'n_neighbors': np.arange(1,30), 
    'metric': ['manhattan', 'euclidean'],
    'weights': ['uniform', 'distance']
    
}
knn_clean = KNeighborsClassifier()
knn_params = cv_params(knn_clean,knn_param_grid)
```

Параметры для RFC

```python
from sklearn.ensemble import RandomForestClassifier
rf_param_grid = {
    'max_depth':[30,40,50,60,70],
    'n_estimators':[200,300,400,600,800]
}
rf_clean = RandomForestClassifier()
rf_params = cv_params(rf_clean,rf_param_grid)
```

```python
import time
#knn
knn = KNeighborsClassifier(**knn_params)
knn.fit(X_train,y_train)
start_time = time.time()
pred_knn = knn.predict(X_test)
pred_knn_train = knn.predict(X_train)
err_test  = np.mean(y_test  != pred_knn)
err_train  = np.mean(y_train  != pred_knn_train)
end_time = time.time()
runtime = end_time - start_time
print('Error on train = ', err_train * 100)
print('Error on test = ', err_test* 100)
print('Runtime = ', runtime)

```

Из результатов заметно, что на обучающей выборке ошибкa меньше, чем на тестовой, но не сильно

```python
#rf

rf = RandomForestClassifier(**rf_params)
rf.fit(X_train,y_train)
start_time = time.time()
pred_rf = rf.predict(X_test)
pred_rf_train = rf.predict(X_train)
err_test  = np.mean(y_test  != pred_rf)
err_train  = np.mean(y_train  != pred_rf_train)
end_time = time.time()
runtime = end_time - start_time
print('Error on train = ', err_train * 100)
print('Error on test = ', err_test* 100)
print('Runtime = ', runtime)
```

Используя случайный лес, ошибка на обучающуй выборке равно 0, а на тестовой ~6.

По результатам видно, что Random Forest дает более высокую точность на обучающей выборке, а на тестовой дает почти такую же, как в предыдущем методе, но работает он гораздо быстрее.  Сравнивая с методом ближайших соседей, в данном случае для задачи классификации лучше использовать Random Forest

Обучим логистическую регрессию

```python
import time
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver="lbfgs", max_iter=1000)
logreg.fit(X_train, y_train)
start_time = time.time()
err_train = np.mean(y_train != logreg.predict(X_train))
err_test  = np.mean(y_test  != logreg.predict(X_test))
end_time = time.time()
runtime = end_time - start_time

print('Ошибка на train = ', err_train)
print('Ошибка на test = ', err_test)
print('Runtime = ', runtime)
```

Полносвязная нейронная сеть c 1-м скрытым слоем

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes = (100,),
                          solver = 'lbfgs', 
                          activation = 'logistic',
                          random_state = 42)
mlp.fit(X_train, y_train)
start_time = time.time()
error_train = np.mean(y_train != mlp.predict(X_train))
error_test  = np.mean(y_test  != mlp.predict(X_test))
end_time = time.time()
runtime = end_time - start_time

print('Ошибка на train = ', error_train)
print('Ошибка на test = ', error_test)
print('Runtime = ', runtime)
```

Подбор гиперпараметров

```python
alphas = np.logspace(-5, 4, 21)

train_err = []
test_err = []

for alpha in alphas :
    mlp = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs' , activation = 'logistic', max_iter=1000, random_state = 42)
    mlp.fit(X_train, y_train)

    y_train_p = mlp.predict(X_train)
    y_test_p = mlp.predict(X_test)
    
    train_err.append(np.mean(y_train != y_train_p))
    test_err.append(np.mean(y_test != y_test_p))

plt.semilogx(alphas, train_err, 'b-o', label = 'train')
plt.semilogx(alphas, test_err, 'r-o', label = 'test')
plt.xlim([np.min(alphas), np.max(alphas)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print('Минимальная ошибка на  train = ', min_train_err)
print('Минимальная ошибка на test = ', min_test_err)
```

```python
alpha_opt = alphas[test_err == min_test_err]
print('Оптимальный параметр alpha = ', alpha_opt)
```

График зависимости ошибки от количества нейронов 

```python
neurons = np.array([i for i in range(1, 600, 20)])

train_err = []
test_err = []

for neuron in neurons:
    mlp = MLPClassifier(alpha = alpha_opt[0], hidden_layer_sizes = (neuron,), 
                              solver = 'lbfgs', activation = 'logistic', max_iter=1000, random_state = 42)
    mlp.fit(X_train, y_train)

    y_train_p = mlp.predict(X_train)
    y_test_p = mlp.predict(X_test)
    
    train_err.append(np.mean(y_train != y_train_p))
    test_err.append(np.mean(y_test != y_test_p))

plt.semilogx(neurons, train_err, 'b-o', label = 'train')
plt.semilogx(neurons, test_err, 'r-o', label = 'test')
plt.xlim([np.min(neurons), np.max(neurons)])
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print('Минимальная ошибка на  train = ', min_train_err)
print('Минимальная ошибка на test = ', min_test_err)
```

```python
neurons_opt = neurons[test_err == min_test_err]
print('Оптимальное количество нейронов = ', neurons_opt)
```
